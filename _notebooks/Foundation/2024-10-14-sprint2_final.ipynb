```python
from random import randint

# Treasures on the island
treasures = [
    {"name": "Gold Coin", "value": 10},
    {"name": "Silver Necklace", "value": 15},
    {"name": "Emerald Gem", "value": 20},
    {"name": "Pearl", "value": 5},
]

# Player and Pirate's scores
player_score = 0
pirate_score = 0
game_over = False
turns = 5  # Total turns for the game

def dig_for_treasure():
    # Simulate digging for treasure (player's action)
    treasure = treasures[randint(0, len(treasures)-1)]
    print(f"\nYou found a {treasure['name']} worth {treasure['value']} points!")
    return treasure['value']

def pirate_action():
    # Simulate pirate's action (pirate also digs)
    treasure = treasures[randint(0, len(treasures)-1)]
    print(f"Pirate found a {treasure['name']} worth {treasure['value']} points!")
    return treasure['value']

def play_game():
    global player_score, pirate_score, turns, game_over
    
    print("Welcome to the Treasure Hunt!")
    
    while not game_over:
        # Player's turn
        print(f"\nTurns remaining: {turns}")
        action = input("\nDo you want to 'dig' for treasure or 'pass' this turn? ").lower()
        
        if action == 'dig':
            player_score += dig_for_treasure()
        elif action == 'pass':
            print("\nYou chose to pass this turn.")
        else:
            print("Invalid input. Please try again.")
            continue
        
        # Pirate's turn (always digs)
        pirate_score += pirate_action()
        
        # Decrease turns and check if the game is over
        turns -= 1
        if turns == 0:
            game_over = True
    
    # End of the game: Display final scores
    print("\nGame Over!")
    print(f"\nYour score: {player_score}")
    print(f"Pirate's score: {pirate_score}")
    
    if player_score > pirate_score:
        print("\nCongratulations! You collected more treasure than the pirate!")
    elif player_score < pirate_score:
        print("\nOh no! The pirate collected more treasure than you!")
    else:
        print("\nIt's a tie! You both collected the same amount of treasure.")

# Start the game
play_game()
```
